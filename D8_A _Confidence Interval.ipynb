```python
from scipy import stats
import pandas as pd
import numpy as np
```

```python
glaxo_df=pd.read_csv('glaxo_df.csv')
glaxo_df.head()
```

```python
glaxo_df.shape

# 1738 rows, 4 columns
```

```python
glaxo_df_ci = stats.norm.interval(0.95,           # 95% confidencial interval
loc = glaxo_df.gain.mean(),                       # gain-->mean
scale = glaxo_df.gain.std()/np.sqrt(1738))        # s/Squar(n)

print( 'Gain at 95% confidence interval is:', np.round(glaxo_df_ci, 4))
```

```python
print(glaxo_df_ci)
```

```python
beml_df=pd.read_csv('beml_df.csv')

beml_df_ci = stats.norm.interval(0.95,
loc=beml_df.gain.mean(),
scale=beml_df.gain.std())



beml_df_ci
```

```python
# Glaxo data frame normal interval
# stats.norm.interval(class interval ,gain-->column -->sample mean , sigma/sqareroot(n) )

glaxo_df_ci=stats.norm.interval(0.95,glaxo_df.gain.mean(),glaxo_df.gain.std()/np.sqrt(1738))
glaxo_df_ci

# Normal interval min,max
```

```python
# Probability
# x(p<=680),loc= mean ,scale=std


stats.norm.cdf(680,loc=711,scale=29)
```

```python
# what is P(697<x<740)

1-stats.norm.cdf(680,loc=711,scale=29)
```

```python
# what is P(697<x<740)

stats.norm.cdf(740,loc=711,scale=29)-stats.norm.cdf(697,loc=711,scale=29)
```
